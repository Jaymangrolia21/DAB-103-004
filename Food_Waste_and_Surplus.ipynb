```python
# find data type, missing values, uniques values with counts
def sniff_modified(df):
    with pd.option_context("display.max_colwidth", 20):
        info = pd.DataFrame()
        info['data type'] = df.dtypes
        info['percent missing'] = df.isnull().sum()*100/len(df)
        info['No. unique'] = df.apply(lambda x: len(x.unique()))
        info['unique values'] = df.apply(lambda x: x.unique())
        return info.sort_values('data type')
```

```python
df.food_type.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Total Count by Food Type")
plt.ylabel('Count of Food Type')
plt.xlabel('Food Type');

for index, value in enumerate(df['food_type'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.show()
```

```python
df.sector.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Total Count by Sector")
plt.ylabel('Count of Sector')
plt.xlabel('Sector');

for index, value in enumerate(df['sector'].value_counts()):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.show()
```

```python
df.groupby('state')['tons_waste'].mean().plot(kind='bar', figsize=(10, 5))

plt.title("Average waste in tons by State")
plt.ylabel('Average waste in tons')
plt.xlabel('State')

plt.show()
```
