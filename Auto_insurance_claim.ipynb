```python
#plot the histogram to see the distribution of the point data.
sns.displot(insurance_data, x="age")
plt.title('Count according to age')
```

```python
sns.heatmap(insurance_data.corr(),cmap="Blues")
plt.title('Incident Types')
```

```python
plt.figure(figsize=(20,10))
plt.title("Check Outlier with boxplot for total claim amount")
insurance_data.boxplot(column=['total_claim_amount']);
```

```python
plt.figure(figsize=[22,10])
sns.displot(insurance_data['total_claim_amount'])
plt.title('Total Claim Amount Distribution Plot')
```

```python
insurance_data['incident_type'].value_counts(normalize=True).plot.pie(startangle=30, autopct='%1.1f%%')
plt.title('Incident Types')
plt.show()
```

```python
insurance_data['insured_sex'].value_counts(normalize=True).plot.pie(startangle=180, autopct='%1.1f%%')
plt.title('Insured Sex')
plt.show()
```

```python
insurance_data['fraud_reported'].value_counts(normalize=True).plot.pie(startangle=90, autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))

insurance_data['auto_make'].value_counts().plot.bar()

plt.show()
```

```python
plt.figure(figsize=(12,5))
incident_type = sns.countplot(data = insurance_data, x = 'incident_severity', hue = 'insured_sex')
incident_type.tick_params(axis='x', rotation=90)
plt.xlabel('Incident Severity Class')
plt.ylabel('Number of Incident')
plt.title('Incident Severity as per Gender')
plt.show()
```

```python
plt.figure(figsize=(12,6))
insurance_data['authorities_contacted'].value_counts().plot.bar()
plt.xlabel('Authorities Types')
plt.ylabel('Number of Accident')
plt.title('Authorities contacted during accident')
plt.show()
```

```python
plt.figure(figsize=(12,6))
insurance_data['incident_state'].value_counts().plot.bar()
plt.xlabel('States')
plt.ylabel('Number of Accident')
plt.title('Most accident happened in different states')
plt.show()
```
